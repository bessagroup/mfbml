```python
import numpy as np


from mfbml.problems.illutrative_problems import mf_Forrester
```

```python
# initialize the problem
problem = mf_Forrester()

# samples
x = np.linspace(0, 1, 1000).reshape(-1, 1)
yh = problem.hf(x)
ylf1 = problem.lf_1(x)
ylf2 = problem.lf_2(x)
ylf3 = problem.lf_3(x)
ylf_factor = problem.lf_factor(x, beta_0=1, beta_1=0.1)
```

```python
#
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.plot(x, yh, label="High fidelity")
ax.plot(x, ylf1, label="Low fidelity 1")
ax.plot(x, ylf2, label="Low fidelity 2")
# ax.plot(x, ylf3, label="Low fidelity 3")
ax.plot(x, ylf_factor, label="Low fidelity factor")
ax.legend()
plt.show()
```

```python
# calculate the correlation
from scipy.stats import pearsonr

print(pearsonr(yh.flatten(), ylf1.flatten()))
print(pearsonr(yh.flatten(), ylf2.flatten()))
print(pearsonr(yh.flatten(), ylf3.flatten()))
print(pearsonr(yh.flatten(), ylf_factor.flatten()))
```
