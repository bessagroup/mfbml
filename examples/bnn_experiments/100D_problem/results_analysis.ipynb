```python
#
import numpy as np
import pandas as pd
import f3dasm
from f3dasm import ExperimentData
```

```python
# read the results
original_exp = ExperimentData.from_file(
    filename='experiments/exp_100d_problem',)
# new results on diff
new_exp = ExperimentData.from_file(
    filename='new_experiments/exp_100d_problem')
```

```python
original_exp
```

```python
new_exp
```

```python
# merge the input_data and output_data into one pd.DataFrame
original_df = pd.concat([original_exp.input_data.data, original_exp.output_data.data],
                        axis=1)
# new results on diff
new_df = pd.concat([new_exp.input_data.data, new_exp.output_data.data],
                   axis=1)
```

```python
# have the mean of the results according to the method
original_df_mean = original_df.groupby('method').mean()
# new results on diff
new_df_mean = new_df.groupby('method').mean()
# standard deviation
original_df_std = original_df.groupby('method').std()
# new results on diff
new_df_std = new_df.groupby('method').std()
```
